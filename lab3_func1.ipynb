```python
def to_cels(f):
    C = (5 / 9) * (F - 32)
    print(f"to celsius: {C}")

f = float(input("Enter number: "))
to_cels(f)
```

```python
def find(heads, legs):
    chicken = int((4 * heads - legs) / 2)
    rabbit  = int(heads - chicken)
    print(f"chickens: {chicken}, rabbits: {rabbit}")

heads = 35
legs = 94
find(heads, legs)
```

```python
def find_prime(numbers):
    for i in numbers:
        if i > 1:
            is_prime = True
            for j in range(2, i): 
                if i % j == 0:
                    is_prime = False
                    break
            if is_prime:
                print(f"Prime number: {i}")

n = int(input("Enter number of elements: "))

numbers = []
for i in range(n):
    numbers.append(int(input(f"Enter number {i + 1}: "))) 

find_prime(numbers)
```

```python
from itertools import permutations

def get_permutations(string):
    perm_list = permutations(string)
    for perm in perm_list:
        print(''.join(perm))

user_input = input("Enter a string: ")
get_permutations(user_input)
```

```python
def reverse_words(text):
    words = text.split(" ")
    words.reverse()
    reversed_str = " ".join(words)
    print(reversed_str)

text = input("Input the string: ")
reverse_words(text)
```

```python
def has_3_3(nums):
    for i in range(len(nums) - 1):
        if nums[i] == nums[i + 1] and nums[i] == 3:
            return True
    return False

nums = []
n = int(input("How many nums: "))
for i in range(n):
    temp = int(input())
    nums.append(temp)
if has_3_3(nums):
    print("True")
else:
    print("False")
```

```python
def spy_game(nums):
    temp = [0, 0, 7]
    for i in range(len(nums)):
        if nums[i] == temp[0]:
            temp.pop(0)
    if not temp:
        return True
    return False

nums = []
n = int(input("How many nums: "))
for i in range(n):
    temp = int(input())
    nums.append(temp)
if spy_game(nums):
    print("True")
else:
    print("False")
```

```python
def Volume_Of_Sphere(r):
    v = (4 / 3) * 3.14159265359 * (r ** 3)
    return v

n = float(input("R = "))
print(Volume_Of_Sphere(n))
```

```python
def Unique(nums):
    new1 = []
    new2 = nums
    new2.sort()
    if len(new2) > 0:  
        new1.append(new2[0])
    for i in range(1, len(nums)):
        if new2[i - 1] != new2[i]:
            new1.append(nums[i])
    return new1

nums = []
n = int(input("How many nums: "))
for i in range(n):
    temp = int(input())
    nums.append(temp)
print(Unique(nums))
```

```python
def Is_It_Palindrom(n):
    m = ""
    for i in range(len(n) - 1, -1, -1):
        m += n[i]
    if n == m:
        print("Yes")
    else:
        print("No")

n = input()
Is_It_Palindrom(n)
```

```python
import random

def Great():
    print("Hello! What is your name?")

def Great2(n):
    print(f"Well, {n}, I am thinking of a number between 1 and 20.")

def Play(m, name):
    win = False
    cnt = 0
    print("If you want to stop the game print \"stop\" without \"")
    while not win:
        print("Take a guess")
        n = input()
        cnt += 1
        if n == "stop":
            print("You lose")
            break
        elif int(n) == m:
            print(f"Good job, {name}! You guessed my number in {cnt} guesses!")
            break
        elif int(n) < m:
            print("Your guess is too low.")
        else:
            print("Your guess is too high.")

Great()
name = input()
Great2(name)
number = random.randint(1, 20)
Play(number, name)
```
